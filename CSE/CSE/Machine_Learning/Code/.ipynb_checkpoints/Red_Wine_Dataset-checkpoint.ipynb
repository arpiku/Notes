```python
import pandas as pd
df = pd.read_csv('winequalityN.csv')
df.head()
```

```python
quality_mapping = {3: 0,4: 1, 5: 2,6: 3,7: 4,8:5}
#one can use mapping funtion in panda to map the data according to this mapping
df.loc[:,"quality"] = df.quality.map(quality_mapping)
df.head()
```

```python
df = df.sample(frac=1).reset_index(drop = True) #Removes the shuffled indices as a column when 'drop is true'
df.head()
```
